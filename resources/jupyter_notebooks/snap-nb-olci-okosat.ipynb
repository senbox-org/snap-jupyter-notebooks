## **SNAP Jupyter demo notebook**
**Using a sequence of various SNAP operators to derive an OLCI L2 water product with a comprehensive set of water quantities.**

This workflow has originally been realized in an XML graph manually built within the Okosat project. In summary, the steps are:

- Read a subset of an S3 OLCI data product using the SNAP 'Read' operator
- Apply an Idepix OLCI pixel classification.
- Generate an appropriate tile cache  for the Idepix node.
- Apply a vicarious calibration of the input radiances using the SNAP 'BandMaths' operator.
- Create a merge product with the calibrated radiances and the Idepix classification, using the SNAP 'Merge' operator.
- Apply the MphChl algorithm on the merge product, using the SNAP 'MphChl' operator.
- Apply the C2RCC algorithm on the merge product, using the SNAP 'c2rcc.olci' operator.
- Generate an appropriate tile cache  for the C2rcc node.
- Derive further water quantities from MphChl and C2rcc, using the SNAP 'BandMaths' operator.
- Do a final merge of all deriverd quantities, using the SNAP 'Merge' operator.
- Write the result to a NetCDF file using the SNAP 'Write' operator.
- Combine all steps (nodes) into a SNAP GPF xml graph, and run the graph using SNAPISTA.

Complexity: advanced

##### ***Replace with re-calibrated bands and include idepix for c2rcc:***

```python
merge = Operator("Merge", source=["Read", "vicarious", "Idepix"])
incl1 = "<include><productId>Idepix</productId><namePattern>pixel_classif_flags</namePattern></include>"
merge.includes = incl1
excl1 = "<exclude><productId>Read</productId><namePattern>Oa.*radiance</namePattern></exclude>"
merge.excludes = excl1
```

```python
g.add_node(operator=merge, node_id="Merge", source=["Read", "vicarious", "Idepix"])
```

##### **Add a 'MphChl' operator to the graph:***

```python
validPixelExpression = '!quality_flags.invalid AND (!quality_flags.land || quality_flags.fresh_inland_water) and not pixel_classif_flags.IDEPIX_CLOUD'
mph_chl = Operator('MphChl', 
                   validPixelExpression=validPixelExpression, 
                   cyanoMaxValue='1000',
                   chlThreshForFloatFlag='500',
                   exportMph='True',
                   applyLowPassFilter='False')
g.add_node(operator=mph_chl, 
           node_id="mphchl",
           source='Merge')
```

##### **Add a 'c2rcc.olci' operator to the graph:***

```python
validPixelExpression = '!quality_flags.invalid AND (!quality_flags.land || quality_flags.fresh_inland_water) and not pixel_classif_flags.IDEPIX_CLOUD'
c2rcc_olci = Operator('c2rcc.olci',
                   validPixelExpression=validPixelExpression,
                   demName='GETASSE30',
                   thresholdCloudTDown865='0.95',
                   outputRhown='False',
                   outputUncertainties='False')
g.add_node(operator=c2rcc_olci, 
           node_id="c2rcc_tobecached",
           source='Merge')
```
