## **SNAP Jupyter demo notebook**
**Using Read, Collocate and Write operators...** 

In summary, this simple workflow contains:

- Read two subsets of an S3 OLCI data product
- Run the SNAP Collocate operator from a SNAP GPF xml graph using SNAPISTA
- Write the result to a NetCDF file

Complexity: simple

##### ***Some information on the Python environment:***

```python
import os
import sys
print("Python version: " + sys.version)

import sysconfig
print("Location of esa_snappy package: " + sysconfig.get_paths()['purelib'] + os.sep + "esa_snappy")
```

##### ***Import esa_snappy and snapista:***

```python
import esa_snappy
import snapista

from snapista import Graph
from snapista import Operator
from snapista import read_file
from snapista import TargetBand
from snapista import TargetBandDescriptors
```

##### ***Set up a graph:***

```python
try:
    g = Graph()
except Exception as ex:
    print("Cannot set up Snapista Graph():", ex)
```

##### ***Add a 'Read' operator to the graph for reading example input:***

```python
input_dir = os.getcwd() + os.sep + 'data'
input_filename_1 = input_dir + os.sep + 'subset_0_of_S3A_OL_1_EFR____20180509T101337_20180509T101637_20180510T150001_0179_031_065_1980_MAR_O_NT_002.nc'
input_filename_2 = input_dir + os.sep + 'subset_1_of_S3A_OL_1_EFR____20180509T101337_20180509T101637_20180510T150001_0179_031_065_1980_MAR_O_NT_002.nc'
g.add_node(operator=Operator("Read", file=input_filename_1), 
           node_id="Read_1")
g.add_node(operator=Operator("Read", file=input_filename_2), 
           node_id="Read_2")
```

##### ***Add a 'Collocate' operator to the graph. Provide the two input products as list. The first product is the reference product:***

```python
g.add_node(operator=Operator('Collocate'), 
           node_id="Collocate" , 
           source=["Read_1", "Read_2"])
```

##### ***Add a 'Write' operator to the graph. Define a NetCDF output file:***

```python
result_filename = os.getcwd() + os.sep + 'results' + os.sep + 'snap_nb_olci_collocate_result.nc'
g.add_node(operator=Operator("Write", file=result_filename, formatName='NetCDF4-BEAM'), 
           node_id="Write", 
           source='Collocate')
```

##### ***View the graph in XML representation:***

```python
g.view()
```

##### ***Save the graph as XML file to disk:***

```python
saved_graph_name = os.getcwd() + os.sep + 'graphs' + os.sep + 'snap_nb_olci_collocate.xml'
g.save_graph(saved_graph_name)
```

##### ***Run the graph via SNAPISTA:***

```python
g.run()
```
