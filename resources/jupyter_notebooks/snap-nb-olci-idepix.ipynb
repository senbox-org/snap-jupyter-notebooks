## **SNAP Jupyter demo notebook**
**OLCI Idepix - Setting up an OLCI Idepix Pixel Classification operator, create and run a graph with default parameters...**

In summary, this simple workflow contains:

- Read a subset of an S3 OLCI data product
- Run the Idepix OLCI pixel classification plugin from a SNAP GPF xml graph using SNAPISTA
- Write the result to a NetCDF file

Complexity: simple

##### ***Some information on the Python environment:***

```python
import os
import sys
print("Python version: " + sys.version)

import sysconfig
print("Location of esa_snappy package: " + sysconfig.get_paths()['purelib'] + os.sep + "esa_snappy")
```

##### ***Import esa_snappy and snapista:***

```python
import esa_snappy
import snapista

from snapista import Graph
from snapista import Operator
from snapista import read_file
from snapista import TargetBand
from snapista import TargetBandDescriptors
```

##### ***Set up a graph:***

```python
try:
    g = Graph()
except Exception as ex:
    print("Cannot set up Snapista Graph():", ex)
```

##### ***Add a 'Read' operator to the graph for reading example input:***

```python
input_dir = os.getcwd() + os.sep + 'data'
input_filename = input_dir + os.sep + 'subset_0_of_S3A_OL_1_EFR____20180509T101337_20180509T101637_20180510T150001_0179_031_065_1980_MAR_O_NT_002.nc'
g.add_node(operator=Operator("Read", file=input_filename), 
           node_id="Read")
```

##### ***Add an 'Idepix.Olci' operator to the graph. Make sure that Idepix OLCI plugin has been installed in SNAP via Plugin Manager:***

```python
g.add_node(operator=Operator('Idepix.Olci'), node_id="Idepix.Olci" ,source='Read')
```

##### ***Add a 'Write' operator to the graph. Define a NetCDF output file:***

```python
result_filename = os.getcwd() + os.sep + 'results' + os.sep + 'snap_nb_olci_idepix_result.nc'
g.add_node(operator=Operator("Write", file=result_filename, formatName='NetCDF4-BEAM'), 
           node_id="Write", 
           source='Idepix.Olci')
```

##### ***View the graph in XML representation:***

```python
g.view()
```

##### ***Save the graph as XML file to disk:***

```python
saved_graph_name = os.getcwd() + os.sep + 'graphs' + os.sep + 'snap_nb_olci_idepix.xml'
g.save_graph(saved_graph_name)
```

##### ***Run the graph via SNAPISTA:***

```python
g.run()
```
